````python
# ============================================================
# AI4ALZHEIMERS - CPU OPTIMIZED VERSION (1-2 hours)
# ============================================================

print("="*60)
print("AI4ALZHEIMERS - CPU OPTIMIZED PIPELINE")
print("="*60)
print("⏱ Estimated time: 1-2 hours on CPU")
print("="*60)

# ===== IMPORTS =====
import os, json, zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import tensorflow as tf
from tensorflow.keras.applications import ResNet50, EfficientNetB0
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import warnings
warnings.filterwarnings('ignore')

from google.colab import drive
drive.mount('/content/drive', force_remount=False)

np.random.seed(42)
tf.random.set_seed(42)

project_dir = '/content/drive/MyDrive/AI4Alzheimers'
print(f"✓ Project: {project_dir}")
print(f"✓ TensorFlow: {tf.__version__}")
print(f"✓ Device: {'GPU' if tf.config.list_physical_devices('GPU') else 'CPU (optimized)'}")

# Create directories
for d in ['data', 'models', 'results', 'reports']:
    os.makedirs(f'{project_dir}/{d}', exist_ok=True)

# ===== DOWNLOAD DATASET =====
print("\n" + "="*60)
print("DATASET SETUP")
print("="*60)

possible_paths = [
    f'{project_dir}/imagesoasis',
    f'{project_dir}/OASIS',
    f'{project_dir}/Data'
]

dataset_dir = None
for path in possible_paths:
    if os.path.exists(path):
        subdirs = os.listdir(path)
        if any('Demented' in d or 'Dementia' in d for d in subdirs):
            dataset_dir = path
            print(f"✓ Found dataset at: {path}")
            break

if dataset_dir is None:
    print("Downloading dataset...")
    os.environ['KAGGLE_USERNAME'] = "shermaine526"
    os.environ['KAGGLE_KEY'] = "3c6f2321c2179d54da4e46f80f514746"

    !pip install -q kaggle
    !kaggle datasets download -d ninadaithal/imagesoasis -p {project_dir}

    zip_path = f'{project_dir}/imagesoasis.zip'
    if os.path.exists(zip_path):
        print("Extracting...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(project_dir)
        os.remove(zip_path)

        for path in possible_paths:
            if os.path.exists(path):
                dataset_dir = path
                break

if dataset_dir is None:
    raise FileNotFoundError("Dataset not found!")

print(f"✓ Dataset: {dataset_dir}")

# ===== LOAD DATA (FASTER - Use subset for CPU) =====
print("\n" + "="*60)
print("LOADING DATA")
print("="*60)

all_subdirs = os.listdir(dataset_dir)
found_dirs = {}

for subdir in all_subdirs:
    if os.path.isdir(os.path.join(dataset_dir, subdir)):
        subdir_lower = subdir.lower()
        if 'non' in subdir_lower and 'dement' in subdir_lower:
            found_dirs[subdir] = 0
        elif 'very' in subdir_lower and 'mild' in subdir_lower:
            found_dirs[subdir] = 0.5
        elif 'mild' in subdir_lower and 'very' not in subdir_lower:
            found_dirs[subdir] = 1
        elif 'moderate' in subdir_lower:
            found_dirs[subdir] = 2

print(f"Found classes: {list(found_dirs.keys())}")

# Load data with sampling for faster training on CPU
data_list = []
SAMPLE_SIZE = 15000  # Use 15k images instead of 86k for faster training

for cdr_class, cdr_value in found_dirs.items():
    class_dir = os.path.join(dataset_dir, cdr_class)
    image_files = [f for f in os.listdir(class_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]

    # Sample proportionally
    if cdr_value == 0:  # Non-demented (most common)
        sample_n = min(10000, len(image_files))
    elif cdr_value == 0.5:  # Very mild
        sample_n = min(3000, len(image_files))
    elif cdr_value == 1:  # Mild
        sample_n = min(1500, len(image_files))
    else:  # Moderate
        sample_n = len(image_files)  # Keep all (only ~500)

    sampled_files = np.random.choice(image_files, size=sample_n, replace=False)

    print(f"  {cdr_class}: {len(sampled_files):,} images (sampled from {len(image_files):,})")

    for img_file in sampled_files:
        data_list.append({
            'filename': img_file,
            'filepath': os.path.join(class_dir, img_file),
            'cdr': cdr_value
        })

df = pd.DataFrame(data_list)
print(f"\n✓ Using {len(df):,} images for faster training")
print(df['cdr'].value_counts().sort_index())

df.to_csv(f'{project_dir}/data/dataset_master.csv', index=False)

# ===== SPLITS =====
print("\n" + "="*60)
print("CREATING SPLITS")
print("="*60)

train_df, temp_df = train_test_split(df, test_size=0.30, stratify=df['cdr'], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.50, stratify=temp_df['cdr'], random_state=42)

print(f"Train: {len(train_df):,}")
print(f"Val: {len(val_df):,}")
print(f"Test: {len(test_df):,}")

train_df.to_csv(f'{project_dir}/data/train_split.csv', index=False)
val_df.to_csv(f'{project_dir}/data/val_split.csv', index=False)
test_df.to_csv(f'{project_dir}/data/test_split.csv', index=False)

# ===== GENERATORS (CPU OPTIMIZED) =====
print("\n" + "="*60)
print("CREATING GENERATORS (CPU OPTIMIZED)")
print("="*60)

train_df['cdr'] = train_df['cdr'].astype(str)
val_df['cdr'] = val_df['cdr'].astype(str)
test_df['cdr'] = test_df['cdr'].astype(str)

# CPU optimized settings
IMG_SIZE = 128  # Smaller for faster processing
BATCH_SIZE = 16  # Smaller batch for CPU
NUM_CLASSES = len(df['cdr'].unique())

print(f"Image size: {IMG_SIZE}x{IMG_SIZE} (smaller for CPU)")
print(f"Batch size: {BATCH_SIZE}")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,  # Less augmentation
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    train_df, x_col='filepath', y_col='cdr',
    target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE,
    class_mode='sparse', shuffle=True, seed=42
)

val_generator = val_test_datagen.flow_from_dataframe(
    val_df, x_col='filepath', y_col='cdr',
    target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE,
    class_mode='sparse', shuffle=False
)

test_generator = val_test_datagen.flow_from_dataframe(
    test_df, x_col='filepath', y_col='cdr',
    target_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE,
    class_mode='sparse', shuffle=False
)

print(f"✓ Generators ready")

# ===== TRAIN 3 MODELS (INSTEAD OF 5) =====
print("\n" + "="*60)
print("TRAINING 3 MODELS (CPU OPTIMIZED)")
print("="*60)
print("⏱ Time per model: ~20-30 min")
print("⏱ Total: ~1-1.5 hours")
print("="*60)

def create_model(base_func, name):
    base = base_func(include_top=False, weights='imagenet',
                     input_shape=(IMG_SIZE, IMG_SIZE, 3))

    # Freeze more layers for faster training
    for layer in base.layers[:int(len(base.layers)*0.8)]:
        layer.trainable = False

    x = GlobalAveragePooling2D()(base.output)
    x = Dense(256, activation='relu')(x)  # Smaller layer
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)
    outputs = Dense(NUM_CLASSES, activation='softmax')(x)

    return Model(base.input, outputs, name=name)

# Only 3 models for CPU
configs = [
    (ResNet50, 'resnet50_1'),
    (EfficientNetB0, 'efficientnet'),
    (ResNet50, 'resnet50_2')
]

trained_models = []

for idx, (base_func, name) in enumerate(configs, 1):
    print(f"\n{'='*60}")
    print(f"MODEL {idx}/3: {name.upper()}")
    print("="*60)

    model = create_model(base_func, name)
    model.compile(
        optimizer=Adam(0.001),  # Higher LR for faster convergence
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    callbacks = [
        ModelCheckpoint(f'{project_dir}/models/{name}.h5',
                       monitor='val_accuracy', save_best_only=True, verbose=1),
        EarlyStopping(monitor='val_loss', patience=3,  # Less patience
                     restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1)
    ]

    # Only 10 epochs instead of 20
    history = model.fit(
        train_generator,
        epochs=10,
        validation_data=val_generator,
        callbacks=callbacks,
        verbose=2  # Less verbose for cleaner output
    )

    val_loss, val_acc = model.evaluate(val_generator, verbose=0)
    print(f"✓ {name}: {val_acc*100:.2f}%")
    trained_models.append(model)

print("\n✓ ALL 3 MODELS TRAINED!")

# ===== ENSEMBLE PREDICTIONS =====
print("\n" + "="*60)
print("ENSEMBLE PREDICTIONS")
print("="*60)

test_generator.reset()
y_true = test_generator.classes

# Ensemble prediction
all_preds = [model.predict(test_generator, verbose=0) for model in trained_models]
all_preds = np.array(all_preds)

mean_preds = np.mean(all_preds, axis=0)
std_preds = np.std(all_preds, axis=0)

y_pred = np.argmax(mean_preds, axis=1)
confidence = np.max(mean_preds, axis=1)
uncertainty = np.mean(std_preds, axis=1)

# Metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"\n{'='*60}")
print("FINAL RESULTS")
print("="*60)
print(f"Accuracy:  {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall:    {recall*100:.2f}%")
print(f"F1-Score:  {f1*100:.2f}%")

# Uncertainty analysis
correct = (y_pred == y_true)
unc_correct = uncertainty[correct]
unc_incorrect = uncertainty[~correct]

print(f"\nUncertainty (correct):   {np.mean(unc_correct):.4f}")
print(f"Uncertainty (incorrect): {np.mean(unc_incorrect):.4f}")

high_risk = uncertainty > 0.15
print(f"High-risk predictions: {high_risk.sum()} ({high_risk.sum()/len(uncertainty)*100:.1f}%)")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['CDR 0', 'CDR 0.5', 'CDR 1', 'CDR 2'],
            yticklabels=['CDR 0', 'CDR 0.5', 'CDR 1', 'CDR 2'])
plt.title('Ensemble Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.savefig(f'{project_dir}/results/confusion_matrix.png', dpi=150)
plt.show()

# Save metrics
metrics = {
    'accuracy': float(accuracy),
    'precision': float(precision),
    'recall': float(recall),
    'f1_score': float(f1),
    'num_models': 3,
    'img_size': IMG_SIZE,
    'total_images': len(df),
    'high_risk_count': int(high_risk.sum())
}

with open(f'{project_dir}/results/test_metrics.json', 'w') as f:
    json.dump(metrics, f, indent=2)

# ===== GENERATE REPORT =====
report = f"""# AI4Alzheimers: Ensemble Deep Learning for Alzheimer's Classification

## Executive Summary

**Results:**
- Ensemble Accuracy: **{accuracy*100:.2f}%**
- F1-Score: **{f1*100:.2f}%**
- High-risk cases: **{high_risk.sum()}** ({high_risk.sum()/len(uncertainty)*100:.1f}%)

**Innovation:**
- 3-model ensemble (ResNet50 x2, EfficientNet-B0)
- Uncertainty quantification via prediction variance
- CPU-optimized for accessibility

## Dataset
- OASIS Dataset: {len(df):,} images (sampled for efficiency)
- CDR 0: {len(df[df['cdr']==0]):,} | CDR 0.5: {len(df[df['cdr']==0.5]):,}
- CDR 1: {len(df[df['cdr']==1]):,} | CDR 2: {len(df[df['cdr']==2]):,}

## Methodology
**Ensemble Architecture:**
- ResNet50 (variant 1)
- EfficientNet-B0
- ResNet50 (variant 2)

**Optimization for CPU:**
- Image size: 128x128 (instead of 224x224)
- 10 epochs with early stopping
- Efficient data augmentation

**Uncertainty Quantification:**
- Uncertainty = std(predictions across ensemble)
- Risk levels: Low (<0.10), Medium (0.10-0.15), High (>0.15)

## Results

| Metric | Score |
|--------|-------|
| Accuracy | {accuracy*100:.2f}% |
| Precision | {precision*100:.2f}% |
| Recall | {recall*100:.2f}% |
| F1-Score | {f1*100:.2f}% |

**Uncertainty Analysis:**
- Mean (correct): {np.mean(unc_correct):.4f}
- Mean (incorrect): {np.mean(unc_incorrect):.4f}

## Clinical Impact
**Workflow Integration:**
1. Low uncertainty → Proceed with diagnosis
2. Medium uncertainty → Additional tests
3. High uncertainty → Manual review ({high_risk.sum()} cases flagged)

## Conclusion
This CPU-optimized system demonstrates:
- Strong performance ({accuracy*100:.2f}% accuracy)
- Safety via uncertainty quantification
- Accessibility (runs on CPU without GPU)

**Submitted for:** Hack4Health AI4Alzheimers Challenge
"""

with open(f'{project_dir}/reports/final_report.md', 'w') as f:
    f.write(report)

# ===== FINAL SUMMARY =====
print("\n" + "="*60)
print("🎉 PROJECT COMPLETE!")
print("="*60)
print(f"\n📊 Results:")
print(f"  Accuracy: {accuracy*100:.2f}%")
print(f"  F1-Score: {f1*100:.2f}%")
print(f"  Models: 3 (CPU-optimized)")
print(f"\n📁 Files:")
print(f"  ✓ 3 models saved")
print(f"  ✓ Metrics: test_metrics.json")
print(f"  ✓ Report: final_report.md")
print(f"\n✓ Ready for submission!")
print("="*60)
````
